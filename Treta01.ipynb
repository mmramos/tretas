```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lat = np.linspace(-24.0, -23.0, 100)
lon = np.linspace(-50.0, -49.0, 100)
LON, LAT = np.meshgrid(lon, lat)
Z = np.cos(LON)*np.sin(LAT)
```

```python
lon_nav = np.linspace(-49.75, -49.15, 100)
lat_nav = [-23.65]*len(lon_nav)
```

```python
plt.figure(figsize=(8,8))
#plt.contourf(LON, LAT, Z, 20, cmap='RdBu_r')
plt.scatter(LON, LAT, s=10.0, c=Z, cmap='RdBu_r')
plt.plot(lon_nav, lat_nav, '.k')
plt.xlabel("LON")
plt.ylabel('LAT')
plt.show()
```

```python
p = 50
print(lat_nav[p],lon_nav[p])
print()
t = [lat_nav[p],lon_nav[p]]


#-----------------------------------------------------------------------#

bat_t = 0
lat_tras = 0
for i in range(len(lat)):
    if lat[i] >= t[0]:
        lat_tras = lat[i]
        bat_t = i
        break
        
#--------------------------------#

bat_f = 0
lat_frente = 0
for i in range(len(lat)):
    if lat[len(lat)-i-1] <= t[0]:
        lat_frente = lat[len(lat)-i-1]
        bat_f = len(lat)-i-1
        break
        
#--------------------------------#

bat_b = 0
lon_baixo = 0
for i in range(len(lon)):
    if lon[i] >= t[1]:
        lon_baixo = lon[i]
        bat_b = i
        break
        
#--------------------------------#

bat_c = 0
lon_cima = 0
for i in range(len(lon)):
    if lon[len(lon)-i-1] <= t[1]:
        lon_cima = lon[len(lon)-i-1]
        bat_c = len(lon)-i-1
        break
        
#--------------------------------#
print('lat =',lat_nav[p],'|lat_tras =',lat_tras,'|lat_frente',lat_frente)
print('lon =',lon_nav[p],'|lon_baixo =',lon_baixo,'|lon_cima',lon_cima)

print(bat_t,bat_f,bat_b,bat_c)
print((bat_t+bat_f+bat_b+bat_c)/4.0)

plt.figure(figsize=(8,8))
#plt.contourf(LON, LAT, Z, 20, cmap='RdBu_r')
plt.scatter(LON, LAT, s=10.0, c=Z, cmap='RdBu_r')
plt.plot(lon_nav, lat_nav, '.k')

plt.plot(lon_baixo, lat_tras, '.g')
plt.plot(lon_cima, lat_tras, '.g')
plt.plot(lon_cima, lat_frente, '.g')
plt.plot(lon_baixo, lat_frente, '.g')

plt.xlabel("LON")
plt.ylabel('LAT')
plt.show()
```
