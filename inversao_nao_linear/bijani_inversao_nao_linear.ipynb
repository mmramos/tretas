$f(x,y) = - \sqrt{xy} sin(x) sin(y)$

```python
def alp(x,y):
    f = - (((x*y)**(0.5))*np.sin(x)*np.sin(y))
    return f
#-----------------------------------------------#

ns = 100
nx = np.array(ns*[np.linspace(-5,5,ns)])
ny = nx.T

nm = np.zeros((ns,ns))
for j in range(ns):
    for i in range(ns):
        nm[j,i] = alp(nx[j,i],ny[j,i])

#print(nx)
plt.imshow(nx)
plt.show()

plt.imshow(ny)
plt.show()

plt.imshow(nm)
plt.show()
```

```python
# DEFINICAO DAS DERIVADAS DE PRIMEIRA E SEGUNDA ORDEM:
def alp(x,y):
    # nonlinear benchmark function for optimization tests
    z = -(x*y)**(0.5) * np.sin(x) * np.sin(y)
    return z
############################################################
def alpx(x,y):
    # derivada da funcao alp em x:    
    a = (-y/2.0)*(x*y)**(-0.5) * np.sin(x) * np.sin(y)
    b = np.sqrt(x*y) * np.cos(x) * np.sin(y)
    z = a - b
    return z
############################################################
def alpy(x,y):
    a = (-x/2.0)*(x*y)**(-0.5) * np.sin(y) * np.sin(x)
    b = np.sqrt(y*x) * np.cos(y) * np.sin(x)
    z = a - b
    return z
############################################################
def alpxx(x,y):
    a = (x**2)/4.0 * (x*y)**(-3.0/2.0) * np.sin(x) * np.sin(y)
    b = (y/2.0) * (x*y)**(-0.5)*np.cos(x) * np.sin(y)
    c = np.sqrt(x*y) * np.sin(x) * np.sin(y)
    z = a - 2.0*b + c
    return z
############################################################
def alpxy(x,y):
    a = y*(x*y)**(-0.5)* np.cos(y)
    b = (x/2.0)*(x*y)**(-0.5) * np.sin(y) + np.sqrt(x*y) * np.cos(y)
    z = -np.sin(x)/2.0 * a - np.cos(x) * b
    return z
############################################################
def alpyx(x,y):
    a = ( np.sin(x) + 2.0*x*np.cos(x) + 2*y*np.cos(y) )
    b = 4.0 * (x*y)**(0.5)
    z = a/b
    return z
############################################################
def alpyy(x,y):
    
    z = ((x**2) * np.sin(x) * (-4 * (y * np.cos(y)) + np.sin(y) + 
                           (4 * (y**2) * np.sin(y))))/(4 * ((x*y)**(3/2)))

    return z
```
