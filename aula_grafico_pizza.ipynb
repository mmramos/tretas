```python
import numpy as np
import matplotlib.pyplot as plt

lit_ = ['Folhelho','$Slurry$','Arenito']
cor_ = ['#3fff00','grey','#ffff3f']

x0 = np.random.randint(0,5,100) # supondo que isso seja a litologia
print(x0)

x1 = list(x0) # transforma em lista

y0 = x1.count(0)
y1 = x1.count(1)
y2 = x1.count(2)
y3 = x1.count(3)
y4 = x1.count(4)

print('contagem =',y0,y1,y2,y3,y4)

Y = [y0,y1,y2,y3,y4]

fig1, ax1 = plt.subplots()
fig1.set_size_inches(5, 5)

ax1.pie(Y)
plt.show()

### fazer usando o loop agora
```

```python
# melhorando o grafico

codigos0 = [
    'ARENITO,ARN,#ffff3f,49', # siliciclástico
    'COQUINA,COQ,#afecec,4', # carbonato
    'CONGLOMERADO,CGL,#ffbe1d,42', # siliciclástico
    'FOLHELHO,FLH,#3fff00,57', # siliciclástico
    'HALITA,HAL,#6dafcd,85', # sal
          ]

codigos_ARN = codigos0[0].split(',')
codigos_COQ = codigos0[1].split(',')
codigos_CGL = codigos0[2].split(',')
codigos_FLH = codigos0[3].split(',')
codigos_HAL = codigos0[4].split(',')

print(codigos_ARN)

nomes = [codigos_ARN[0],codigos_COQ[0],codigos_CGL[0],codigos_FLH[0],codigos_HAL[0]]
cores = [codigos_ARN[2],codigos_COQ[2],codigos_CGL[2],codigos_FLH[2],codigos_HAL[2]]

#---------------------------------------------------------#

fig1, ax1 = plt.subplots()
fig1.set_size_inches(5, 5) # tamanho da figura

ax1.pie(Y, colors=cores, labels=nomes, 
        
        autopct='%5.0f%%', # posição e formato dos números 
        shadow=True, # sombreado
        startangle=90, # angulo inicial em 90º
        pctdistance=0.7) # distância do centro

#ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

### fazer usando o loop agora
```
