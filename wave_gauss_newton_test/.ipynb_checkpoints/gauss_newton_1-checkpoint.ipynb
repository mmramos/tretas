```python
# importando as bibliotecas
import matplotlib.pyplot as plt # graficos
import pandas as pd # armazenamento de dados
import numpy as np # numerica

##################################################################

# classe para pegar os valores de densidade e RHOB e transform=a-los em S.I. (sistema internacional)
class lagemar_log_modeling:
    
    def __init__(self):
        pass
    
    #=============================================================#
    
    def load_RHOB(self,data,SI=False): # Load RHOB
        if SI:
            data2 = data.copy() 
        else:
            data2 = []
            for i in range(len(data)): # gcm3 para kgm3
                data2.append(data[i]*1000.0)
                
        self.rhob_si = data2
        
    #________________________________________#
    
    def load_VP(self,data,SI=False): # Load VP
        cfvp = 3.2808*(10**(-6))
        if SI:
            data2 = data.copy() 
        else:
            data2 = []
            for i in range(len(data)): # us/ft para m/s
                data2.append(304800/data[i])
                
        self.vp_si = data2
        
##################################################################


# função para fazer os gráficos (2 tracks apenas)
def graficos_2(prof,prop1,prop2,label1 = 'RHOB',label2 = 'DTP'):

    ### Graficos
    fig, (ax1, ax2) = plt.subplots(1,2) # 2 cruvas = 2 subplots
    fig.set_size_inches(7, 5) # tamanho da figura | polegadas

    #_____________________________________________________#

    ax1.plot(prop1,prof,'-b',linewidth = 1)
    ax1.invert_yaxis()
    ax1.grid()
    ax1.set_title(label1, fontsize=16)

    #_____________________________________________________#

    ax2.plot(prop2,prof,'-k',linewidth = 1)
    ax2.invert_yaxis()
    ax2.grid()
    ax2.set_title(label2, fontsize=16)

    plt.show()
```

```python
### Importando os dados

# arquivo .las com os dados (caminho)
url = 'DP1.las' 

# upload das curvas de profundidade ('PROF'), densidade ('RHOB') e sônico ('DTP')
cabecalho = ['PROF','RHOB','DTP'] 

# posição das curvas no track
curvas = [0,8,10] 

# importando as curvas via pandas
dado_raw = pd.read_csv(url, header=43,usecols=curvas, sep='\s+',names=cabecalho)

# delimitando o intervalo de amostragem no pandas (2700 a 3000 m)
dado_raw = dado_raw.loc[(dado_raw['PROF'] >= 2700) & (dado_raw['PROF']  < 3000)]

# removendo dummies das curvas 'DTP' e 'RHOB' para facilitar o calculo
# OBS: os dados localizados em dado_raw agora estão localizados em DP1
DP1 = dado_raw.loc[(dado_raw['DTP']  != -999.00000) &
                    (dado_raw['RHOB']  != -999.00000)]

#-------------------------------------------------------------------------------------#
### Graficos
graficos_2(DP1['PROF'],DP1['RHOB'],DP1['DTP'])
```

```python
# modificando os dados para o S.I.

# inicializando a classe
dado = lagemar_log_modeling()

# executando o upload dos dados de 'RHOB' na forma de lista
# OBS: A classe não suporta dados em pandas, portanto precisa ser convertido
dado.load_RHOB(list(DP1['RHOB']))

# executando o upload dos dados de 'DTP' na forma de lista
dado.load_VP(list(DP1['DTP']))

# VP é a variável que contém as velocidades do poço no S.I.
VP  = dado.vp_si

# DEN é a variável que contém as densidades do poço no S.I.
DEN = dado.rhob_si

#-----------------------------------------#

# Graficos de 'DEN' e 'VP'
graficos_2(DP1['PROF'],DEN,VP,'DEN','VP')
```

### Equações:

<br>$V_c = \sqrt{ \frac{Y}{\rho} }$ | materiais mais rígidos (Ellis e Singer)
<br>
<br>$V_c ^2 = \frac{Y}{\rho} \frac{1 - \mu}{(1 + \mu)(1 - 2 \mu)}$ | caso geral
<br>
    
<br>$V_c = \sqrt{ \frac{Y}{\rho} \frac{1 - \mu}{(1 + \mu)(1 - 2 \mu)} }$ | caso geral (modificado)

<br>Onde:
- $V_c$ é a velocidade (ondas compressionais).
- $\rho$ é a densidade.
- $Y$ é o módulo de Young.
- $\mu$ é o coeficiente de poisson.

______________________________________________________________________________________
### Derivadas Parciais:

<br>$\frac{\partial V_c}{\partial \mu} = \frac{A}{B \times C}$

<br>$A = (\mu - 2) \mu Y$
<br>$B = \rho(2 \mu^2 + \mu - 1)^2$
<br>$C = V_c = \sqrt{ \frac{Y}{\rho} \frac{1 - \mu}{(1 + \mu)(1 - 2 \mu)} }$

______________________________________________________________________________________

<br>$\frac{\partial V_c}{\partial Y} = \frac{\sqrt{ \frac{D Y}{\rho} }}{2Y}$

<br>$D = \frac{1 - \mu}{(1 + \mu)(1 - 2 \mu)}$

______________________________________________________________________________________
### Componentes do Gauss-Newton:

<br>$\overline{\dot{\mu}} = \frac{A}{B \times C}$ para cada elemento de $\rho _i$

<br>$\overline{\dot{Y}} = \frac{\sqrt{ \frac{D Y}{\rho _i} }}{2Y}$ para cada elemento de $\rho _i$

<br>$\overline{\overline{J}} = [\overline{\dot{\mu}},\overline{\dot{Y}}]$ | Jacobiano

<br>$\overline{r} = V_c(i)  - f( \rho _i)$ | vetor de resíduos

______________________________________________________________________________________
### Gauss-Newton atualização das constantes:

$C^{j+1} = C^{j} - (\overline{\overline{J}} ^T \overline{\overline{J}})^{-1} \overline{\overline{J}}^{T} \overline{r}$




```python
# formulação no python:
# OBS:
# rho = densidade
# Vc  = velocidade compressional
# u   = coeficiente de poisson
# Y   = módulo de Young

# Lita com valores (slide 14)
# https://www.slideshare.net/MRINALJYOTIADHYAPOK/stress-68255928

# Equação original no caso geral
def ellis_singer(rho,u=0.24,Y=22.0*(10**9)):
    Vc = np.sqrt( (Y/rho)*( (1-u)/( (1+u)*(1-2*u) + 0.00001 ) ) )
    return Vc

# Equação original modificada para visualização dos resultados (sem a raiz quadrada)
def ellis_singer2(rho,u=0.24,Y=22.0*(10**9)):
    #Vc = np.sqrt( (Y/rho)*( (1-u)/( (1+u)*(1-2*u) ) ) *-1.0 ) # comentado por ter raiz quadrada (original)
    Vc = ( (Y/rho)*( (1-u)/( (1+u)*(1-2*u) ) ))
    return Vc

#------------------------------------------------------#
# primeira parcial em relação ao coeficiente de poisson
def ellis_u(rho,u=0.24,Y=22.0*(10**9)):
    A = (u-2)*u*Y
    B = rho*(( ((2*u)**2) + u - 1)**2)
    C = np.sqrt( (Y/rho)*( (1-u)/( (1+u)*(1-2*u) ) ) )
    
    dVc = A/(B*C)
    return dVc
    
# segunda parcial em relação ao módulo de Young
def ellis_Y(rho,u=0.24,Y=22.0*(10**9)):
    D = (1 - u)/( (1 + u)*(1 - (2 * u)) )
    dVc = np.sqrt( ((D*Y)/rho)/(2*Y) )
    return dVc

#------------------------------------------------------#            
    
# Teste do modelo simples
print('teste simples do modelo',ellis_singer(DEN[0]))
```

```python
# Aplicação do modelo

# exemplo de coeficientes (coeficiente de poisson e modulo de Young para o arenito)
#c0[0] = coeficiente de Poisson
#c0[1] = modulo de Young
c0 = [0.24,22.0*(10**9)] # u , Y para o arenito


# determinação do número de variáveis para a profundidade
n = len(DEN)

# transformando a lista de constantes em arrays para permitir operações em algebra linear (numpy)
c02 = np.array(c0)

du = [] # lista contendo valores das derivadas parciais do coeficiente de poisson
dY = [] # lista contendo valores das derivadas parciais do módulo de Young
r  = [] # lista contendo valores de resíduos

Vc_novo = [] # lista que contém os valores de V

for i in range(n):
    # aplicando as funções com o i-ésimo valor de DEN, c0[0], c0[1]
    du.append(ellis_u(DEN[i],c0[0],c0[1])) # derivada parcial em relação ao coeficiente de poisson
    dY.append(ellis_Y(DEN[i],c0[0],c0[1])) # derivada parcial em relação ao módulo de Young
    r.append(VP[i] - ellis_singer(DEN[i],c0[0],c0[1])) # resíduos
    
# definindo a matriz Jacobiana e o vetor de resíduos    
J = np.array([du,dY]).T; print('dimensões da matriz J:',np.shape(J))
r = np.array(r).T; print('dimensões do vetor r:',np.shape(r))

# a produto da matriz transposta por ela mesma
JJ = np.dot(np.transpose(J),J);print('dimensões da matriz JJ:',np.shape(JJ))

# inversa da matriz JJ
JI = np.linalg.inv(JJ);print('dimensões da matriz JI:',np.shape(JI))
    
# produto da inversa de JI com a transposta de J
JT = np.dot(JI,np.transpose(J));print('dimensões da matriz JT:',np.shape(JT))

# produto de JT com o vetor de resíduos
JR = np.dot(JT,r);print('dimensões da matriz JR:',np.shape(JR))

# atualização dos coeficientes
C1 = c02 - JR;print('novos valores de u e Y:',C1)
```

```python
# incorporando as iterações:

# numero de iteracoes a serem realizadas
iteracoes = 100

# chamando novamente os coeficientes
c0 = [0.24,22.0*(10**9)]

# determinação do número de variáveis para a profundidade
n = len(DEN)

# transformando a lista de constantes em arrays para permitir operações em algebra linear (numpy)
c02 = np.array(c0)

# loop com as iterações
for j in range(iteracoes):

    du = []
    dY = []
    r  = []
    VP_novo = []
    for i in range(n):
        du.append(ellis_u(DEN[i],c0[0],c0[1]))
        dY.append(ellis_Y(DEN[i],c0[0],c0[1]))
        r.append(VP[i] - ellis_singer(DEN[i],c0[0],c0[1]))

    J = np.array([du,dY]).T
    r = np.array(r).T

    JJ = np.dot(np.transpose(J),J)

    JI = np.linalg.inv(JJ)

    JT = np.dot(JI,np.transpose(J))
    JR = np.dot(JT,r)
    
    #!!AQUI!! atualização dos coeficientes
    c02 = c02 - JR
    #print('valores atualizados:',c02)

print('constantes obtidas (fim da iteração):',c02)
```

```python
# incorporando as atualizações no modelo:

n_VP = [] # novos valores de Vc (Velocidade Compressional)
VP2 = []  # velocidade compressional ao quadrado
for i in range(n):
    n_VP.append(ellis_singer2(DEN[i],c02[0],c02[1]))
    VP2.append(VP[i]**2)
    
#---------------------------------------------------------------------#

graficos_2(DP1['PROF'],n_VP,VP2,'$VP_n$','$VP^2$')
```

```python
# Teste de valores: Modulo de Young e Coeficiente de Poisson

Y = np.linspace(0.1,0.3,100) # range de valores de Y
u = np.linspace(0.1,0.3,100) # range de valores de u 

MR = np.array([[0.0]*100]*100) # definindo uma matriz de zeros (100x100)

for k in range(100):
    for j in range(100):
    
        r = []
        for i in range(len(DEN)):
            # calculo de resíduos
            r.append(VP[i] - ellis_singer(DEN[i],u[j],Y[k]))
            
        # definindo r como array (vetor)
        r = np.array(r)
        
        # phi = erro ao quadrado dividido pela amostragem
        phi = np.dot(r,r.T)/len(DEN)
        
        # incorporando os valores de phi na matriz
        MR[k,j] = phi

plt.imshow(MR,origin='lower')
plt.colorbar()
plt.show()
```
